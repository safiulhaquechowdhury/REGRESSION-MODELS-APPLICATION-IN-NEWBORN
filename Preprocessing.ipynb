```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv ('babies.csv')
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
mean_gestation = dataset['gestation'].mean()
# Fill missing values in 'gestation' column with the mean
dataset['gestation'].fillna(mean_gestation, inplace=True)
```

```python
dataset.isnull().sum()
```

```python
mean_age = dataset['age'].mean()
mean_height = dataset['height'].mean()
mean_weight = dataset['weight'].mean()
mean_smoke = dataset['smoke'].mean()
```

```python
dataset['age'].fillna(mean_age, inplace=True)
dataset['height'].fillna(mean_height, inplace=True)
dataset['weight'].fillna(mean_weight, inplace=True)
dataset['smoke'].fillna(mean_smoke, inplace=True)
```

```python
dataset.isnull().sum()
```

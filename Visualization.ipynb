```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('babies(Preprocessed).csv')
```

```python
dataset
```

```python
X = dataset.drop('bwt', axis=1)
y = dataset['bwt']
```

# Heatmap

```python
correlation_matrix = dataset.corr()
plt.figure(figsize=(9, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# Histograms

```python
import matplotlib.pyplot as plt

# List of features
features = ['age', 'gestation', 'parity', 'height', 'weight', 'smoke', 'bwt']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(10, 5))

# Flatten the 2D array of subplots
axes = axes.flatten()

# Plot histograms for each feature
for i, feature in enumerate(features):
    axes[i].hist(dataset[feature], bins=20, edgecolor='black')
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

# Density Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of features
features = ['age', 'gestation', 'parity', 'height', 'weight', 'smoke', 'bwt']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(10, 5))

# Flatten the 2D array of subplots
axes = axes.flatten()

# Plot density plots for each feature
for i, feature in enumerate(features):
    sns.kdeplot(data=dataset, x=feature, ax=axes[i], fill=True)
    axes[i].set_title(f'Density Plot of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Density')

# Adjust layout
plt.tight_layout()
plt.show()
```

# Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of features
features = ['age', 'gestation', 'parity', 'height', 'weight', 'smoke', 'bwt']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(10, 5))

# Flatten the 2D array of subplots
axes = axes.flatten()

# Plot transparent box plots for each feature
for i, feature in enumerate(features):
    sns.boxplot(x=feature, data=dataset, ax=axes[i], boxprops=dict(alpha=0.5))
    axes[i].set_title(f'Box Plot of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Value')

# Adjust layout
plt.tight_layout()
plt.show()
```

# Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of features
features = ['age', 'gestation', 'parity', 'height', 'weight', 'smoke', 'bwt']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 8))

# Flatten the 2D array of subplots
axes = axes.flatten()

# Plot violin plots for each feature
for i, feature in enumerate(features):
    sns.violinplot(x=feature, data=dataset, ax=axes[i], inner="quartile")
    axes[i].set_title(f'Violin Plot of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Value')

# Adjust layout
plt.tight_layout()
plt.show()
```
